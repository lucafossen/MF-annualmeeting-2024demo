```python
import pandas as pd
import os
import pickle
os.chdir('../')
```

```python
articles_selection = pd.read_csv('data/test_set_articles.csv')
all_data = pd.read_csv('data/df_2018_2023.csv')
with open('data/recommendations.pkl', 'rb') as file:
    recommendations = pickle.load(file)
```

```python
print("Keys in recommendations:")
print(recommendations[0].keys())
print("Number of dictionaries in recommendations:", len(recommendations))

print("\nColumns in articles_selection:")
print(articles_selection.columns.tolist())
print("Number of rows in articles_selection:", len(articles_selection))

print("\nColumns in all_data:")
print(all_data.columns.tolist())
print("Number of rows in all_data:", len(all_data))
```

```python
# Create a set of the union of all related articles in articles_selection and recommendations
related_articles_union = set()

# Iterate over each row in articles_selection
for index, row in articles_selection.iterrows():
    # Convert the string representation of the list to an actual list
    related_articles = eval(row['related_articles'])
    # Update the set with the related articles from the current row
    related_articles_union.update(related_articles)

# Iterate over each item in the recommendations list
for recommendation in recommendations:
    # Update the set with the recommended articles from the current item
    related_articles_union.update(recommendation['recommendations'])

# Display the resulting set
related_articles_union
```

```python
# Filter the full dataset to include only rows with uuids present in the related_articles_union set
filtered_data = all_data[all_data['uuid'].isin(related_articles_union)]

# Display the filtered DataFrame
filtered_data
```

```python
# Export the filtered data to a CSV file
filtered_data.to_csv('articles_big_dataset.csv', index=False)
```
